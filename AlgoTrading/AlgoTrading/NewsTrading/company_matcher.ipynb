```python
import pandas as pd
import pickle
file_path = 'MCAP.xlsx'
df = pd.read_excel(file_path)
symbols = df.iloc[:,1]
company_names = df.iloc[:,2]
company_dict = {name: symbol for name, symbol in zip(company_names, symbols)}
with open('company_data.pkl', 'wb') as f:
    pickle.dump(company_dict,f)
```
